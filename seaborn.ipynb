# Seaborn

## importar libreria y traer datos de ejemplo

```python
import seaborn as sns
import matplotlib.pyplot as plt

datos = sns.load_dataset("tips")
print(datos.head())
```

## Grafico de dispersion (scatterplot)

```python
sns.scatterplot(data=datos, x="total_bill", y="tip")
```

## Histograma (histplot)

```python
sns.histplot(data=datos, x="total_bill", bins=20)
```

## Diagrama de caja (boxplot)

```python
sns.boxplot(data=datos, x="day", y="total_bill")
```

## Grafico de barras (barplot)

```python
sns.barplot(data=datos, x="day", y="total_bill")
```

## Grafico de violin (violinplot)

```python
sns.violinplot(data=datos, x="day", y="total_bill")
```

## Añadir titulo y etiqueta a los ejes

plt.title("titulo de grafico")
plt.xlabel("etiqueta eje x")
plt.ylabel("etiqueta eje y")

```python
sns.scatterplot(data=datos, x="total_bill", y="tip")
plt.title("Relacion entre el total de la cuenta y la propina")
plt.xlabel("Total de la cuenta")
plt.ylabel("Propina")
```
